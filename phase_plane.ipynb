```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('file.txt')['start'].values
```

```python
data
```

```python
starts = [0] + list(np.where(data == 'start')[0])
```

```python
finish = np.where(data == 'finish')[0]
```

```python
after_gap = np.where(data == 'aftergap')[0]
after_tau = np.where(data == 'aftertau')[0]
```

```python
def get_phase(idx, data):
    list_phase = list(data[(starts[idx]+1):after_gap[idx]]) + list(data[(after_gap[idx]+2):after_tau[idx]]) + list(data[(after_tau[idx]+2):finish[idx]])
    N = len(list_phase)
    tau = data[after_tau[idx]+1]
    data_theta = [float(list_phase[:N][i]) for i in range(1, N, 2)]
    data_R = [float(list_phase[:N][i])for i in range(0, N, 2)]   
    x = [ data_R[i] * np.cos(data_theta[i]) for i in range(len(data_R))]
    y = [ data_R[i] * np.sin(data_theta[i]) for i in range(len(data_R))]
    return data_theta, data_R, tau, x, y
```

```python
t, r, tau, x, y = get_phase(10, data)
```

```python
plt.title('Changes in Amplitude')
plt.plot(r, 'o')
plt.xlabel('time')
plt.ylabel('Amplitude')
```
